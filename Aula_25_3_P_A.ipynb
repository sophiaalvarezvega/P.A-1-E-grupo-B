```
#Programa que recebe 4 notas, calcula a média e diz se o aluno foi aprovado.

print("As notas não podem ser maior que 10")
nota1 = float(input("Digite a nota 1 bimestre: "))
nota2 = float(input("Digite a nota 2 bimestre: "))
nota3 = float(input("Digite a nota 3 bimestre: "))
nota4 = float(input("Digite a nota 4 bimestre: "))

resultado = (nota1 + nota2 + nota3 + nota4) / 4

if resultado >= 7:
  print(f"Sua nota foi {resultado}, você foi aprovado")

elif resultado <= 5:
  print(f"Você terá que fazer a recuperação")

else:
  print(f"Você tirou {resultado}, foi reprovado")
```
