```
#Você foi contratado para desenvolver um sistema simples de caixa eletrônico.

#O sistema deverá funcionar através de um menu interativo utilizando a estrutura de repetição while.

#O programa deverá iniciar com um saldo inicial de $1000.00 e permitir que o usuário escolha diferentes operações até decidir sair do sistema.

#Funcionalidades

# ver saldos
# depositar dinheiro
# sacar dinheiro
# sair

# regras

# ver saldo
# ao escolher essa opção, o sistema deverá mostrar o saldo atual da conta.

# depositar dinheiro
# o usuário deverá informar um valor para depósito.
# o valor digitado será somado ao saldo.

# sacar dinheiro
# o usuário deverá informar o valor do saque

# o sistema devera verificar:

# se houver suficiente - realizar o saque
# se não houver saldo suficiente - mostrar mensagem de saldo insuficiente

# Sair

# o programa deverá encerrar apenas quando o usuario escolher a opção.


saldo = 1000.0
opcao = 0

while opcao != 4:

    print("=====~CAIXA ELETRÔNICO~=====")
    print("1 - Ver saldo")
    print("2 - Depositar dinheiro")
    print("3 - Sacar dinheiro")
    print("4 - Sair")

    opcao = int(input("Escolha uma opção: "))

    # Ver saldo
    if opcao == 1:
        print(f"Seu saldo atual é: R$ {saldo}")

    # Depositar dinheiro
    elif opcao == 2:
        deposito = float(input("Digite o valor do depósito: R$ "))

        if deposito > 0:
            saldo += deposito
            print("Depósito")
            print(f"Novo saldo: R$ {saldo}")
        else:
            print("Valor inválido")

    # Sacar dinheiro
    elif opcao == 3:
        saque = float(input("Digite o valor do saque: R$ "))

        if saque <= saldo:
            saldo -= saque
            print("Saque realizado")
            print(f"Novo saldo: R$ {saldo}")
        else:
            print("Saldo insuficiente")
    elif opcao == 4:
        print("Encerrando")

    # Opção inválida
else:
        print("Opção inválida, tente novamente.")
```
