```
#Você foi contratado para desenvolver um sistema simples de caixa eletrônico.

#O sistema deverá funcionar através de um menu interativo utilizando a estrutura de repetição while.

#O programa deverá iniciar com um saldo inicial de $1000.00 e permitir que o usuário escolha diferentes operações até decidir sair do sistema.

#Funcionalidades

# ver saldos
# depositar dinheiro
# sacar dinheiro
# sair

# regras

# ver saldo
# ao escolher essa opção, o sistema deverá mostrar o saldo atual da conta.

# depositar dinheiro
# o usuário deverá informar um valor para depósito.
# o valor digitado será somado ao saldo.

# sacar dinheiro
# o usuário deverá informar o valor do saque

# o sistema devera verificar:

# se houver suficiente - realizar o saque
# se não houver saldo suficiente - mostrar mensagem de saldo insuficiente

# Sair

# o programa deverá encerrar apenas quando o usuario escolher a opção.


saldo = 1000.0
opcao = 0

while opcao != 4:

    print("=====~CAIXA ELETRÔNICO~=====")
    print("1 - Ver saldo")
    print("2 - Depositar dinheiro")
    print("3 - Sacar dinheiro")
    print("4 - Sair")

    opcao = int(input("Escolha uma opção: "))

    # Ver saldo
    if opcao == 1:
        print(f"Seu saldo atual é: R$ {saldo}")

    # Depositar dinheiro
    elif opcao == 2:
        deposito = float(input("Digite o valor do depósito: R$ "))

        if deposito > 0:
            saldo += deposito
            print("Depósito")
            print(f"Novo saldo: R$ {saldo}")
        else:
            print("Valor inválido")

    # Sacar dinheiro
    elif opcao == 3:
        saque = float(input("Digite o valor do saque: R$ "))

        if saque <= saldo:
            saldo -= saque
            print("Saque realizado")
            print(f"Novo saldo: R$ {saldo}")
        else:
            print("Saldo insuficiente")
    elif opcao == 4:
        print("Encerrando")

    # Opção inválida
else:
        print("Opção inválida, tente novamente.")
```

```
#Jogo menu inteligente

#Criar um sistema que funciona até o usuário escolher sair.

#Menu

# 1 - Par/Ímpar
# 2 - Positivo/Negativo
# 3 - Maior
# 4 - Sair

while True:
    print("=== MENU ===")
    print("1 - Descobrir se é Par ou Ímpar")
    print("2 - Descobrir se é Positivo ou Negativo")
    print("3 - Descobrir qual número é Maior")
    print("4 - Sair")

    opcao = input("Escolha uma opção: ")

    # Par ou Ímpar
    if opcao == "1":
        numero = int(input("Digite um número: "))

        if numero % 2 == 0:
            print("Esse número é PAR!")
        else:
            print("Esse número é ÍMPAR!")

    # Positivo ou Negativo
    elif opcao == "2":
        numero = float(input("Digite um número: "))

        if numero > 0:
            print("Esse número é POSITIVO!")
        elif numero < 0:
            print("Esse número é NEGATIVO!")
        else:
            print("O número é ZERO!")

    # Maior número
    elif opcao == "3":
        n1 = float(input("Digite o primeiro número: "))
        n2 = float(input("Digite o segundo número: "))

        if n1 > n2:
            print(f"O maior número é {n1}")
        elif n2 > n1:
            print(f"O maior número é {n2}")
        else:
            print("Os dois números são iguais!")

    # Sair
    elif opcao == "4":
        print("Programa encerrado. Até mais!")
        break

    # Opção inválida
    else:
        print("Opção inválida! Tente novamente.")
```

```
#Urna eletrõnica

# Votar candidato A
# Votar candidato B
# Votar candidato C
# Votar candidato D
# Votar branco
# Votar nulo
# Encerrar a votação

# O sistema deve repetir até encerrar a votação.
# Contar votos de cada candidato
# Contar votos brancos e nulos
# Ao encerrar, mostrar:

# Total de votos
# Total de votos por candidato
# Total votos brancos
# Total votos nulos
# Vencedor da eleição

# Regras extras

# Se houver empate entre candidatos: mostrar eleição empatada
# Se não hopuver votos válidos, mostrar: "nenhum candidato recebeu votos"

# Urna Eletrônica

# Urna Eletrônica

# Urna Eletrônica

# Quantidade de votos
a = 0
b = 0
c = 0
d = 0
branco = 0
nulo = 0

while True:

    print("\n===== URNA ELETRÔNICA =====")
    print("1 - Candidato A")
    print("2 - Candidato B")
    print("3 - Candidato C")
    print("4 - Candidato D")
    print("5 - Branco")
    print("6 - Nulo")
    print("7 - Encerrar votação")

    voto = input("Escolha seu voto: ")

    if voto == "1":
        a += 1
        print("Voto no candidato A confirmado!")

    elif voto == "2":
        b += 1
        print("Voto no candidato B confirmado!")

    elif voto == "3":
        c += 1
        print("Voto no candidato C confirmado!")

    elif voto == "4":
        d += 1
        print("Voto no candidato D confirmado!")

    elif voto == "5":
        branco += 1
        print("Voto em branco confirmado!")

    elif voto == "6":
        nulo += 1
        print("Voto nulo confirmado!")

    elif voto == "7":

        total = a + b + c + d + branco + nulo

        print("\n===== RESULTADO FINAL =====")
        print(f"Total de votos: {total}")
        print(f"Candidato A: {a}")
        print(f"Candidato B: {b}")
        print(f"Candidato C: {c}")
        print(f"Candidato D: {d}")
        print(f"Brancos: {branco}")
        print(f"Nulos: {nulo}")

        # Descobrir vencedor
        maior = max(a, b, c, d)

        if maior == 0:
            print("Nenhum candidato recebeu votos.")

        elif (
            (a == maior and b == maior) or
            (a == maior and c == maior) or
            (a == maior and d == maior) or
            (b == maior and c == maior) or
            (b == maior and d == maior) or
            (c == maior and d == maior)
        ):
            print("A eleição terminou empatada!")

        elif a == maior:
            print("O vencedor foi o candidato A!")

        elif b == maior:
            print("O vencedor foi o candidato B!")

        elif c == maior:
            print("O vencedor foi o candidato C!")

        else:
            print("O vencedor foi o candidato D!")

        break

    else:
        print("Opção inválida!")
```
