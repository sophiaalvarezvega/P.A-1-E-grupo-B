```
#Jogo da adivinhação

#O usuário terá 03 chances para adivinhar um número de 0 a 10.
#Se o número secreto for maior ou menor que o palpite o programa irá imprimir uma mensagem informando isso.
#Informar se o palpite está quente ou frio.

import random

numero_secreto = random.randint(0, 10)

for tentativa in range(3):
    palpite = int(input("Digite um número entre 0 e 10: "))

    if palpite == numero_secreto:
        print("Você acertou!")
        break

    if palpite < numero_secreto:
        print("O número secreto é maior.")
        diferenca = numero_secreto - palpite
    else:
        print("O número secreto é menor.")
        diferenca = palpite - numero_secreto

    if diferenca <= 2:
        print("Quente")
    else:
        print("Frio")

else:
    print("Você perdeu!")
    print("O número secreto era", numero_secreto)
```
