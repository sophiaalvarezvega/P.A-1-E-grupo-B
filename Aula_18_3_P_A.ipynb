```
#Operadores aritméticos


#Fazer um programa onde o usuario digita dois numeros
# Saída

#Entrada

a = int(input("Digite um numero: "))
b = int(input("Digite um outro numero"))

#Processamento

soma = a + b
subtração = a - b
divisão = b/a
div_int = b//a
resto_da_divisao_é = b % a
multiplicação = a * b
potenciação = a ** b

#Saída

print(f"O resultado de {a} mais {b} é igual a {soma}")
print(f"O resultado de {a} menos {b} é igual a {subtração}")
print(f"O resultado de {b} dividido por {a} é igual a {divisão}")
print(f"O resultado inteiro de {b} dividido por {a} é igual a {div_int} e seu resto é {resto_da_divisao_é}")
print(f"O resultado de {a} multiplicado por {b} é igual a {multiplicação}")
print(f"O resultado de {a} elevado a {b} é igual a {potenciação}")
```

```
#Fazer um programa que converta horas em minutos e horas em segundos

horas = int(input("Digite as horas: "))

horas_pra_minutos = horas * 60

print(f"O resultado da conversão de {horas} horas pra minutos é igual a {horas_pra_minutos} minutos")
```

```
#Fazer um programa que converta km em metros

km = int(input("Digite os km: "))

km_pra_metros = km * 1000

print(f"O resultado da conversão de {km} km pra metros é igual a {km_pra_metros} metros")
```

```
#Fazer um programa que converta km em milhas

km = int(input("Digite os km: "))

km_pra_milhas = km * 0.6214

print(f"O resultado da conversão de {km} km pra milhas é igual a {km_pra_milhas} milhas")
```
